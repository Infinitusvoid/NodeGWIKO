```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from PIL import Image, ImageOps

# Manually specified list of words with their frequencies
word_frequencies = {
    'Transformers': 100,
    'self-attention': 80,
    'feedforward layers of the residual blocks': 90,
    'pretraining dataset': 70,
    'CIFAR-10': 60,
    'transferring from one modality to another': 60,
    'extract features across tokens': 50,
    'zeroshot generalization': 55,
    'data-rich modality': 45,
    'bitstrings': 40,
    'parse mathematical expressions': 35,
    'linear probing': 30,
    'ListOps': 30,
    'frozen attention layers': 25,
    'linear regression': 20,
    'finetuning layernorm': 20,
}

# Define a function to generate a color for each word
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    import random
    return "hsl({}, 100%, 50%)".format(random.randint(0, 255))

# Create a word cloud object
wordcloud = WordCloud(
    width=1024,
    height=1024,
    background_color='black',  # Black background
    color_func=color_func
).generate_from_frequencies(word_frequencies)

# Convert to image and add a border
image = wordcloud.to_image()
bordered_image = ImageOps.expand(image, border=5, fill='yellow')

bordered_image.save("generate_image.png")


bordered_image
```
