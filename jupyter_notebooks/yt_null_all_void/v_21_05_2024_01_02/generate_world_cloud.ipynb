```python
import torch
from PIL import Image
import numpy as np

def load_image(image_path):
    """Load an image from file and convert it to a PyTorch tensor."""
    image = Image.open(image_path).convert('RGB')
    image_tensor = torch.from_numpy(np.array(image)).float()
    return image_tensor

def save_image(image_tensor, output_path):
    """Save a PyTorch tensor as an image file."""
    image_np = image_tensor.numpy().astype(np.uint8)
    image = Image.fromarray(image_np)
    image.save(output_path)

def change_color(image_tensor, target_color, new_color):
    """
    Change all pixels of a target color to a new color and set all other pixels to black.

    Parameters:
    - image_tensor: PyTorch tensor of shape (H, W, 3)
    - target_color: List or tuple of the RGB color to change (e.g., [255, 0, 0] for red)
    - new_color: List or tuple of the new RGB color (e.g., [0, 255, 0] for green)
    """
    target_color = torch.tensor(target_color, dtype=torch.float32)
    new_color = torch.tensor(new_color, dtype=torch.float32)
    black_color = torch.tensor([0, 0, 0], dtype=torch.float32)
    
    mask = (image_tensor == target_color).all(dim=-1)
    
    # Set all pixels to black
    image_tensor[:, :] = black_color

    # Set target color pixels to new color
    image_tensor[mask] = new_color

    return image_tensor


def contains_color(tensor, color):
    """
    Check if the specified color is present in the tensor.

    Parameters:
    tensor (torch.Tensor): A tensor of shape (1090, 1930, 3).
    color (list or tuple): A list or tuple representing the RGB values of the color [r, g, b].

    Returns:
    bool: True if the color is present, False otherwise.
    """
    # Convert color to a tensor
    color_tensor = torch.tensor(color, dtype=tensor.dtype, device=tensor.device)

    # Check for the presence of the color in the tensor
    is_present = (tensor == color_tensor).all(dim=-1).any().item()

    return is_present

# Example usage:
# Load the image
image_path = output_folder+"generate_image_.png"
output_path = output_folder+"generate_image_torch.png"
image_tensor = load_image(image_path)

# Define the target color and the new color
target_color = [100, 0, 8]  # Red color
new_color = [255, 255, 255]     # Green color

# Change the color in the image tensor
modified_image_tensor = change_color(image_tensor, target_color, new_color)

# Save the modified image
save_image(modified_image_tensor, output_path)


# Check if the image contains the target color
if contains_color(image_tensor, target_color):
    print("The image contains the target color.")
else:
    print("The image does not contain the target color.")
```

```python
# Example usage:
# Load the image
image_path = output_folder+"generate_image_.png"
output_path = output_folder+"generate_image_torch.png"
image_tensor = load_image(image_path)


def contains_color(tensor, color):
    """
    Check if the specified color is present in the tensor.

    Parameters:
    tensor (torch.Tensor): A tensor of shape (1090, 1930, 3).
    color (list or tuple): A list or tuple representing the RGB values of the color [r, g, b].

    Returns:
    bool: True if the color is present, False otherwise.
    """
    # Step 1: Convert color to a tensor
    color_tensor = torch.tensor(color, dtype=tensor.dtype, device=tensor.device)

    # Step 2: Compare each pixel in the tensor with the specified color
    # This will create a boolean tensor of the same height and width where each element is True
    # if the corresponding pixel in the original tensor matches the specified color
    comparison = tensor == color_tensor  # comparison shape will be (1090, 1930, 3)
   
    # Step 3: Check if all three channels (R, G, B) match using the all(dim=-1) method
    all_channels_match = comparison.all(dim=-1)  # all_channels_match shape will be (1090, 1930)
    #print(all_channels_match)
    # Step 4: Check if any of the pixels in the tensor match the specified color
    is_present = all_channels_match.any().item()

    return is_present

color_r = 0.0
while True:
    # Define the target color and the new color
    target_color = [255.0, 255.0, color_r]  # Red color
    new_color = [255, 255, 255]     # Green color

    # Change the color in the image tensor
    modified_image_tensor = change_color(image_tensor, target_color, new_color)

    # Save the modified image
    save_image(modified_image_tensor, output_path)

    print(image_tensor.size())

    # Check if the image contains the target color
    if contains_color(image_tensor, target_color):
        print("target_color : ", target_color)
        print("The image contains the target color.")
    else:
        print("target_color : ", target_color)
        print("The image does not contain the target color.")
        break
    color_r += 1
    
```
